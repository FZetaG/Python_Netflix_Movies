```python
# Importar pandas y matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Leer Netflix CSV como un DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
#Filtrar solamente el tipo de show que sea 'movie'
netflix_subset=netflix_df[netflix_df['type'] == 'Movie']
netflix_subset.head(10)
```

```python
# Filtrar peliculas unicamente lanzadas en la decada de los 90
movies_1990 = netflix_subset[(netflix_subset['release_year'] >= 1990) & (netflix_subset['release_year'] <= 1999)]
movies_1990.head(10)
```

```python
#histograma con la distribución de las peliculas en los 90
plt.hist(movies_1990['duration'],edgecolor='black')
plt.title('Distribución de la duración de las peliculas en los 90')
plt.xlabel('duración en minutos')
plt.ylabel('numero de peliculas')
plt.grid(alpha=0.5)
plt.show()

duration=100
```

```python
#filtrar las peliculas de la decada de los 90 solamente por peliculas de accion
action_movies_1990 = movies_1990[movies_1990['genre'] == 'Action']
action_movies_1990.head(10)
```

```python
#Contar cantidad de peliculas de menos de 90 minutos con iterrow
short_movie_count = 0

for label, row in action_movies_1990.iterrows():
    if row['duration'] < 90:
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count= short_movie_count
        
print(short_movie_count)
```

```python
#Agrupar por genero y agregar por el promedio de duracion de las peliculas de la decada de los 90, por cada año
movies_1990.pivot_table(values='duration', index='genre',columns='release_year',margins=True, fill_value=0)
```

```python
#Obtener un ranking de las 10 peliculas con mayor duración de la decada de 1990
movies_1990.loc[:,['director','title','country','release_year','duration']].sort_values('duration',ascending=False).head(10)
```

```python
#Hallar las peliculas con mayor duracion de toda la base de datos y ordernarla de manera descendente
all_movies_rank=netflix_subset.loc[:,['director','title', 'country','release_year','duration']].sort_values('duration',ascending=False)
print(all_movies_rank.head(10))
```

```python
#Filtrar las peliculas segun los países que aparecen dentro de las 10 primeras en termino de duración
countries_to_filter = ['United States', 'Egypt', 'India', 'United Kingdom']

all_movies_rank_filtered=all_movies_rank[all_movies_rank['country'].isin(countries_to_filter)][['country','duration']]
```

```python
#elaborar una grafica descriptiva por pais en base a los resultados anteriores
import seaborn as sns

sns.violinplot(
    x = "country",
    y = "duration",
    data=all_movies_rank_filtered
)
plt.title('Violin Plot de duración de la película por país')
plt.xlabel('País')
plt.ylabel('Duración')
plt.xticks(rotation=45)  
plt.grid(True)
plt.tight_layout()
plt.show()


```
