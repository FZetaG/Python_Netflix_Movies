```python
# Importar pandas y matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Leer Netflix CSV como un DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
#Filtrar solamente el tipo de show que sea 'movie'
netflix_subset=netflix_df[netflix_df['type'] == 'Movie']
netflix_subset.head(10)
```

```python
# Filtrar peliculas unicamente lanzadas en la decada de los 90
movies_1990 = netflix_subset[(netflix_subset['release_year'] >= 1990) & (netflix_subset['release_year'] <= 1999)]
movies_1990.head(10)
```

```python
#histograma con la distribución de las peliculas en los 90
plt.hist(movies_1990['duration'])
plt.title('Distribución de la duracion de las peliculas en los 90')
plt.xlabel('duración en minutos')
plt.ylabel('numero de peliculas')
plt.show()

duration=100
```

```python
#filtrar las peliculas de la decada de los 90 solamente por peliculas de accion
action_movies_1990 = movies_1990[movies_1990['genre'] == 'Action']
action_movies_1990.head(10)
```

```python
#Contar cantidad de peliculas de menos de 90 minutos con iterrow
short_movie_count = 0

for label, row in action_movies_1990.iterrows():
    if row['duration'] < 90:
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count= short_movie_count
        
print(short_movie_count)
```

```python
#Agrupar por genero y agregar por el promedio de duracion de las peliculas de la decada de los 90, por cada año
movies_1990.pivot_table(values='duration', index='genre',columns='release_year',margins=True, fill_value=0)

```
